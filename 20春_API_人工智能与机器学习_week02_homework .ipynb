## 人脸识别产品说明

### 实现了什么？  

- 通过运用人脸识别技术，实现了识别人物是否有戴眼镜、表情、年龄、性别的识别。  

### 附加加值？  

- 将表头转换成中文，使结果更具可读性。  

### 服务于？    

###### 这款产品可以服务于两个方面。    

- 征兵活动  
  - 近视：通过对眼镜的佩戴情况的识别，可以轻松筛选过滤掉部分近视青年。  
  - 适龄：筛选出符合征兵年龄的青年。  
  - 心理健康：推测青年的心理状况是否健康，抗压能力如何。  
- 学生管理  
  - 座位安排：依据学生们的近视状况快速安排且合理地座位表，将近视的同学排在教室前面。  
  - 心理辅导：有些同学学业压力大（尤其是高三学生）或家庭不和睦，容易造成严重的心理问题。借助该产品能够帮助老师察觉和了解学生的基本心理状况，做出及时的心理疏导。
